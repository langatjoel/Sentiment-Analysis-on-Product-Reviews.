## Data Exploaration and Cleaning
The main aim was to explore and clean the Amazon Reviews dataset to prepare it for modelling.
The dataset was checked for missing values,incorrect data types and overall structure to ensure data quality

## Data Quality
- The data has minimal missing values,which were handled by removing or imputing the missing entries.
- The data types were consistent and ready for analysis.
- No major duplicates were found that could affect the results.

## General Observations
- Most of the customers' ratings are **4 or 5 stars**, showing positive reviews in general.
- **1 or 2 stars** ratings usually correspond to negative reviews.
- The dataset features thousands of distinct products and reviews.

## Review Text Insights
- Positive reviews often involve short and direct phrases like *“Great product!”* or *“I Loved it.”*
- Negative reviews are normally longer, with complaints described in detail.
- Frequent words:
- Positive: **great, **love**, **good**, **excellent**
- Negative: **bad**, **poor**, **disappointed**, **waste**

## Insights of the Features
- A new feature `review_length` was added to measure the number of words in each review.
- Poorer ratings are associated with longer reviews,indicating that dissatisfied customers tend to give longer reviews.

## Overall Summary

Now, the Amazon Reviews dataset is clean and prepared for feature engineering. There is a clear relationship between text, rating, and sentiment:
- **Positive sentiment** mostly 4–5 star ratings 
- **Neutral sentiment**  3 stars 
- **Negative sentiment** 1–2 star ratings 
These findings will lead into the next stage of the project to build features and prepare the data for modelling.
